```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from matplotlib.pyplot import Line2D
import math
import matplotlib.animation as animation

cross_color = (0, 0, 0)
bg_color = (.9,.9,.9)
circle_color_list = {
    'red':    (1,0,0),
    'green':  (0,1,0),
    'blue':   (0,0,1),
    'yellow': (1,1,0),
    'cyan':   (0,1,1),
    'magenta':(1,0,1),
    'black':  (0,0,0),
    'white':  (1,1,1),
}
#右下角作者
def draw_text():
    ax1.text(8, 0.01, 'a-chien.blogspot.com', fontsize=5)

#繪製中央的十字
def draw_cross():
    line=Line2D((4.8,5.2),(5,5), color = cross_color )
    ax1.add_line(line)
    line=Line2D((5,5),(4.8,5.2), color = cross_color )
    ax1.add_line(line)

#繪製12個圓圈
def draw_circles():
    circle_num = 12
    circle_dist = 4
    for i in range(circle_num):
        x_pos = 5+math.cos( math.pi*2/circle_num*i ) * circle_dist
        y_pos = 5+math.sin( math.pi*2/circle_num*i ) * circle_dist
        cir=Circle(( x_pos ,y_pos ), radius=.6)
        ax1.add_patch(cir)

#設定初始的畫布        
def init():
    ax1.set_xlim(0, 10)      #x軸長度
    ax1.set_ylim(0, 10)      #y軸長度
    ax1.set_aspect('equal')  #x和y等比例
    ax1.axis('off')          #不顯示座標軸
    draw_text()              #繪製右下角版權 
    draw_cross()             #繪製中央十字
    draw_circles()           #繪製圓圈
    
def animate(frame, *fargs):
    circle_color = fargs
    print(circle_color)
    circles = ax1.patches #取得所有在ax1的圓圈
    
    for item in circles:
        item.set_facecolor(circle_color)
        item.set_edgecolor(circle_color)
        
    #將特定圓圈的顏色設定與背景相同 = 圓圈消失
    circles[ frame%12].set_facecolor(bg_color)
    circles[ frame%12].set_edgecolor(bg_color)
    

for color_name in circle_color_list:
    fig = plt.figure( figsize=(3, 3), dpi=100)
    ax1 = fig.add_subplot(111)
    fig.patch.set_facecolor( bg_color )

    anim=animation.FuncAnimation(fig = fig,
                                 #func = animate(circle_color_list[color_name] ),
                                 func = animate,
                                 fargs = circle_color_list[color_name],
                                 init_func = init,
                                 frames = 12,
                                 interval = 100,
                                 blit = False)
    anim.save( color_name+'.gif',
              writer = 'pillow',
              dpi = 100,
              savefig_kwargs={'transparent': True, 'facecolor': bg_color }
             )
```
